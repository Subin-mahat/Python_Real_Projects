# Q.4. Create a new column in the dataframe "age_group" . This column should contain the age group the respondent belongs to . The age groups are 18-25, 25-40, 40-55 and 55+. The dataframe should look like this after the column creation:

```python
survey.head(2)
```

```python
data = survey.copy()                               # To make a copy of dataframe
```

```python
data.head(2)
```

```python
data.insert(10, 'age_group', survey.age)
```

```python
data.head()
```

```python
data1 = data[(data.age_group >= 18) & (data.age_group < 25)]                    # filtering with 'and' operator
```

```python
data1.head(2)
```

```python
data1.age_group.unique()
```

```python
data1['age_group'] = '18-25'                     # Assigning the values to a column
```

```python
data1
```

```python
data2 = data[(data.age_group >= 25) & (data.age_group < 40)]                     # Filtering with 'and' operator
```

```python
data.head(2)
```

```python
data2.age_group.unique()
```

```python
data2['age_group'] = '25-40'
```

```python
data2.head(2)
```

```python
data3 = data[(data.age_group > 40) & (data.age_group < 55)]                  # Filtering with 'and' operator
```

```python
data3.head(2)
```

```python
data3.age_group.unique()
```

```python
data3['age_group'] = '40-55'
```

```python
data3.head(2)
```

```python
data4 = data[(data.age_group >= 55)]                                              # Filtering on condition base
```

```python
data4.head(2)
```

```python
data4['age_group'] = '55+'
```

```python
data4.head(2)
```

```python
data = pd.concat([data1, data2, data3, data4])                           # For concatenation of the dataframes
```

```python
data
```
